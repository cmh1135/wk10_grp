Machine learning algorithms requires variables that are in numeric form. In cases where catagorical data is presented it must be translated to numeric data (integers) in order to be operated on. One hot coding converts the random integers in to binary variables (0,1) with 1 being assigned to the index variable allowing the integers to be used without having a ordinal relationship. This makes for better predictions and performance.
